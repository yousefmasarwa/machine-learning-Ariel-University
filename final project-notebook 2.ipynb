```python
print("Q1")
def dec_to_bin(x):
    return str(bin(x)[2:])

def dec_to_oct(x):
    return str(oct(x)[2:])
def dec_to_hex(x):
    return str(hex(x)[2:])

##main##
x= int(input("Enter your num: "))
print("base binary is: ",dec_to_bin(x))
print("base octal is: ",dec_to_oct(x))
print("base hex is: ",dec_to_hex(x))
```
